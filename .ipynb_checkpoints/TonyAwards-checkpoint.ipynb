Now, I want to build plots that show the dividing line between before Tony season and after. These are instructive to see the overall effect of a play receiving a Tony on different metrics.

```python
def plot_by_tony_award_and_y_val(tony_str, tony, y_val, y_val_str, action_text = "Receiving", filename = ""):
    f, axarr = plt.subplots(4, 4)
    f.set_figheight(15)
    f.set_figwidth(15)
    f.suptitle("""{} Before and After {} the Best {} Tony Award
    (vertical red line denotes Tony award season)""".format(y_val_str, action_text, tony_str), fontsize = 20)
    for i, play in enumerate(tony.keys()):
        play_data = broadway[broadway['Name'] == play]
        axarr[i // 4, i % 4].plot(play_data['Full'], play_data[y_val])
        axarr[i // 4, i % 4].axvline(x=parsedt(tony[play]), c='r')
        play_str = play[:30] + "..." if len(play) > 30 else play
        axarr[i // 4, i % 4].set_title("{} ({})".format(play_str, parsedt(tony[play]).year))
        axarr[i // 4, i % 4].axis('off')

    f.subplots_adjust(hspace=0.3)
    plt.savefig(filename)
    plt.show()
```

These are pretty to look at and all--but can we actually see the change in average weekly gross income from the month before receiving the Tony Award to the month after? I decide to ignore the change in capacity as, above, it looks quite chaotic, and there are several plays that have 100% capacity for a significant period of time.

```python
import numpy as np
def gross_increase_bar_by_tony_award(tony_str, tony, status_text = "Receiving", filename=""):
    fig, ax = plt.subplots()
    plt.title("""{}s' Gross Income Percentage Increase The Month After {} a Tony Award""".format(tony_str, status_text), fontsize = 20)
    y = []
    for i, play in enumerate(tony.keys()):
        play_data = broadway[broadway['Name'] == play]
        before = play_data[play_data['Full'] < parsedt(tony[play])].tail(4)['Gross'].mean()
        after = play_data[play_data['Full'] > parsedt(tony[play])].head(4)['Gross'].mean()
        y.append(after / before)
    fig.set_figwidth(15)
    fig.set_figheight(10)
    ax.barh(np.arange(16), (np.array(y) - 1) * 100)
    ax.set_yticks(np.arange(16))
    ax.set_yticklabels(tony.keys())
    ax.invert_yaxis()  # labels read top-to-bottom
    plt.savefig(filename)
    plt.show()
    return np.array(y) - 1
```

```python
musical_increases = gross_increase_bar_by_tony_award("Musical", musical_winners, filename = "musical_winners_bar.jpg")
```

```python
play_increases = gross_increase_bar_by_tony_award("Play", play_winners, filename = "play_winners_bar.jpg")
```

```python
musical_nominee_increases = gross_increase_bar_by_tony_award("Musical Nominee", musical_nominees, "Being Nominated For", "musical_nominees_bar.jpg")
```

```python
play_nominee_increases = gross_increase_bar_by_tony_award("Play Nominee", play_nominees, "Being Nominated For", "play_nominees_bar")
```

And, finally, we look at the average increase in aggregate for both musicals and plays.

```python
y = np.array([musical_increases.mean(), play_increases[play_increases > -100].mean(),
             musical_nominee_increases[musical_nominee_increases > -100].mean(), play_nominee_increases[play_nominee_increases > -100].mean()])
fig, ax = plt.subplots(figsize=(15, 5))
plt.title("Average Percentage Increase in Gross Income the Month After Tony Award Season")
plt.bar(np.arange(4), (y * 100))
ax.set_xticks(np.arange(4))
ax.set_xticklabels(('Musical Winners', 'Play Winners', 'Musical Nominees (sample)', 'Play Nominees (sample)'))
plt.savefig("final_increases.jpg")
plt.show()
```

Winning a Tony award seems to be worth a 15 - 20% increase for play productions for both Musicals and Plays! This is a pretty interesting finding. 

```python
plays = broadway[broadway['Type'] == "Play"]
musicals = broadway[broadway['Type'] == "Musical"]

all_play_winners = plays[plays['Name'].isin(play_winners.keys())]
all_plays_after_2000 = plays[plays['Full'] >= datetime.datetime(2000, 1, 1)]

play_winner_percentage = all_play_winners['Gross'].sum() / all_plays_after_2000['Gross'].sum()
```

```python
all_musical_winners = musicals[musicals['Name'].isin(musical_winners.keys())]
all_musicals_after_2000 = musicals[musicals['Full'] >= datetime.datetime(2000, 1, 1)]

musical_winner_percentage = all_musical_winners['Gross'].sum() / all_musicals_after_2000['Gross'].sum()
```

```python
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Best Musicals', 'Others'
sizes = [musical_winner_percentage, 1 - musical_winner_percentage]
explode = (0.2, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.title("All Broadway musicals income since 2000")
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.savefig("gross_percentage_winners_musicals.png")
plt.show()
```

```python
labels = 'Best Plays', 'Others'
sizes = [play_winner_percentage, 1 - play_winner_percentage]
explode = (0.2, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.title("All Broadway play income since 2000")
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.savefig("gross_percentage_winners_plays.png")
plt.show()
```

```python
run_times_musical_winners = all_musical_winners.groupby("Name").apply(lambda x: x['Full'].max() - x['Full'].min())
run_times_play_winners = all_play_winners.groupby("Name").apply(lambda x: x['Full'].max() - x['Full'].min())
gross_musical_winners = all_musical_winners.groupby("Name").sum()['Gross']
gross_play_winners = all_play_winners.groupby("Name").sum()['Gross']
```

```python
non_winning_musicals = all_musicals_after_2000[~all_musicals_after_2000['Name'].isin(musical_winners)]
non_winning_plays = all_plays_after_2000[~all_plays_after_2000['Name'].isin(play_winners)]

run_times_musicals = non_winning_musicals.groupby("Name").apply(lambda x: x['Full'].max() - x['Full'].min())
run_times_plays = non_winning_plays.groupby("Name").apply(lambda x: x['Full'].max() - x['Full'].min())
gross_musicals = non_winning_musicals.groupby("Name").sum()['Gross']
gross_plays = non_winning_plays.groupby("Name").sum()['Gross']
```

```python
plt.scatter(run_times_musicals.apply(lambda x: x.days), gross_musicals / 1000000, label="Non-winners", c = "#e5e5e5")
plt.scatter(run_times_musical_winners.apply(lambda x: x.days), gross_musical_winners / 1000000, label="Best Musical winners")
plt.legend()
plt.title("Musicals: Run time vs Total Gross Income")
plt.ylabel("Total gross income (millions of USD)")
plt.xlabel("Run time (days)")
plt.savefig("runtime_musicals.png")
plt.show()
```

```python
plt.scatter(run_times_plays.apply(lambda x: x.days), gross_plays / 1000000, label="Non-winners", c = "#e5e5e5")
plt.scatter(run_times_play_winners.apply(lambda x: x.days), gross_play_winners / 1000000, label="Best Play winners")
plt.legend()
plt.title("Plays: Run time vs total gross income")
plt.ylabel("Total gross income (millions of USD)")
plt.xlabel("Run time (days)")
plt.savefig("runtime_plays.png")
plt.show()
```
